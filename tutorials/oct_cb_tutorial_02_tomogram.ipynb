## There are several different ways to compute a tomogram using the reconstruct library (oct.reconstruct.tomogram).
#### Technique 1: Direct variable assignment without class instance

This technique is most like MATLAB, but does not take avantage of all the initialization that can be taken advantage of thanks to object oriented programming.

```python
out = Tomogram().reconstruct(ch1=data.ch1,
                         ch2=data.ch2,
                         bg1=data.bg1,
                         bg2=data.bg2,
                         chirp=data.chirp,
                         dispersion=data.dispersion,
                         settings=data.reconstructionSettings)
```

#### Technique 2: Create tomogram instance -> direct variable assigment

Using this technique, we can keep the class in memory and continue to process other frames without needing to reinitialize and variables or memory space, important for GPU processing

```python
tom = Tomogram(mode='heterodyne')

out = tom.reconstruct(ch1=data.ch1,
                         ch2=data.ch2,
                         bg1=data.bg1,
                         bg2=data.bg2,
                         chirp=data.chirp,
                         dispersion=data.dispersion,
                         settings=data.reconstructionSettings)
# Load new frame
data.loadFringe(frame=1)

out = tom.reconstruct(ch1=data.ch1,
                         ch2=data.ch2,
                         bg1=data.bg1,
                         bg2=data.bg2,
                         chirp=data.chirp,
                         dispersion=data.dispersion,
                         settings=data.reconstructionSettings)
```

#### Technique 3 (preferred): Create tomogram instance -> `data` variable assigment

This method is the preferred way to process data is the Dataset loader was used. This method automatically grabs all the required variables and settings from the `data` object and reconstructs the tomogram using it.

```python
tom = Tomogram(mode='heterodyne')
out = tom.reconstruct(data=data)
```

## What is in "out"?

The processed tomogram resides in an output dictionary. This output dictionary can also include the linearized and dispersion compensated k space variables as well as the stokes vectors. These outputs are decided based on the variable 'processState' within `data.reconstructionSettings['processState']`

```python
for key,val in out.items():
    if not (out[key] is None):
        print('Dictionary key ', key, ' : ', out[key].shape, out[key].dtype)
    else:
        print('Dictionary key ', key, ' : ', 'None' , 'None')
```

To have the kspace fringes outputted, add ' +kspace'  to the process state. To initiate stokes vector creation add '+ps'. Since there are many polarization outputs that require stokes vectors (dop,ret,theta), we use the common string "ps" as the process state indicator

```python
data.reconstructionSettings['processState'] = 'kspace+ps'
tom = Tomogram(mode='heterodyne')
out = tom.reconstruct(data=data)

for key,val in out.items():
    if not (out[key] is None):
        print('Dictionary key ', key, ' : ', out[key].shape, out[key].dtype)
    else:
        print('Dictionary key ', key, ' : ', 'None' , 'None')
```

## Let's make sure this is true by trying a few differet scenarios:

```python
# Can reconstruct single channel data with no other inputs
tom = Tomogram()
outi = tom.reconstruct(ch1=data.ch1)
```

```python
# Can reconstruct single channel data with other inputs
tom = Tomogram()
out = tom.reconstruct(ch1=data.ch1, bg1=data.bg1, settings=data.reconstructionSettings)
```

```python
# Can reconstruct two channel data with other inputs
tom = Tomogram()
out = tom.reconstruct(ch1=data.ch1, 
                      ch2 =data.ch2, 
                      bg1=data.bg1, 
                      bg2 = data.bg1, 
                      settings=data.reconstructionSettings)
```

```python
# Finally we can include chirp and dispersion
tom = Tomogram()
outf = tom.reconstruct(ch1=data.ch1, 
                      ch2 =data.ch2, 
                      bg1=data.bg1, 
                      bg2 = data.bg1, 
                      chirp = data.chirp,
                      dispersion = data.dispersion,
                      settings=data.reconstructionSettings)
```
