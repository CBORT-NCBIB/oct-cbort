## First let's process some data and write to file

Normally the only outputs that will be saved are struct, Angio as well as the ps contrasts: dop, ret and theta. 

However, there are some intermediate processed data type that if explicitl called in processFrameRange, they will **write** their outputs to .h5 files. MGH filetype has advantages for uint8 data, such as memory mapping, however for other datatypes it remains complicated.

These are:
1. tomo (complex tomograms without spectral binning)
2. kspace (k space fringes after disp/chirp compensation)
3. stokes (stokes vectors)
4. oa (optic axes 1,2,3)

```python
d = os.path.join(module_path,'examples/data/4_BL_Benchtop_Phantom_struct_angio_ps/')
data = Load(directory = d)

processer = Post()
processer.processFrameRange(data, startFrame=1, endFrame=1, procState='tomo+angio+struct+ps+kspace+stokes+oa', 
                            writeState=True,
                            procAll=False, 
                            hold=False, 
                            holdState='struct', 
                            rotCartesian=False)
```

### If Tomograms were saved during processing, they can also be loaded using the loadh5 function

```python
f1= os.path.join(d,'[PSTest][Phantom][12-01-2020_09-53-12].tomch1.h5')
f2= os.path.join(d,'[PSTest][Phantom][12-01-2020_09-53-12].tomch2.h5') 

# Note the dataset name or "dataStr" will be the same name as the data key ie: tomch1 
tomch1 = dataset.loadH5(f1, frame=1, dataStr='tomch1')
tomch2 = dataset.loadH5(f2, frame=1, dataStr='tomch2')

# Take a look at the structure (note contrast has not been adjusted as in previous example)
plt.imshow(np.log10(abs(tomch1**2+tomch2**2)))
```

### If k-space fringes were saved during processing, they can also be loaded using the loadh5 function

```python
f1= os.path.join(d,'[PSTest][Phantom][12-01-2020_09-53-12].k1.h5')
f2= os.path.join(d,'[PSTest][Phantom][12-01-2020_09-53-12].k2.h5') 

# Note the dataset name or "dataStr" will be the same name as the data key ie: tomch1 
k1 = dataset.loadH5(f1, frame=1, dataStr='k1')
k2 = dataset.loadH5(f2, frame=1, dataStr='k2')

tomch1 = cp.fft.fft(k1*processer.tom.fringeWindow,n=processer.tom.settings['numZOut'],axis=0)
tomch2 = cp.fft.fft(k2*processer.tom.fringeWindow,n=processer.tom.settings['numZOut'],axis=0)
tomch1 = cp.roll(tomch1, +  processer.tom.settings['demodReverseIndex'], axis=0)
tomch2 = cp.roll(tomch2, +  processer.tom.settings['demodReverseIndex'], axis=0)
# Take a look at the structure (note contrast has not been adjusted as in previous example)
plt.imshow(np.flipud(np.log10(abs(tomch1**2+tomch2**2))))
```

### If raw stokes vectors were saved during processing, they can also be loaded using the loadh5 function

```python
f1= os.path.join(d,'[PSTest][Phantom][12-01-2020_09-53-12].sv1.h5')
f2= os.path.join(d,'[PSTest][Phantom][12-01-2020_09-53-12].sv2.h5') 
sv1 = dataset.loadH5(f1, frame=1, dataStr='sv1')
sv2 = dataset.loadH5(f2, frame=1, dataStr='sv2')
```

### We can now continue to process the data from these stokes vectors

```python
# With symmetric mode
ps = Polarization()
out = ps.reconstruct(sv1=sv1,sv2=sv2)
plt.imshow(out['ret'])
```

```python
# Settings will be the defaults from the Metadata object since no settings are loaded here. 
# In order to change something, ie. maxRet, simply re-process after editting ps.settings.
ps.settings['maxRet'] = 80
out = ps.reconstruct(sv1=sv1,sv2=sv2)
plt.imshow(out['ret'])
```
