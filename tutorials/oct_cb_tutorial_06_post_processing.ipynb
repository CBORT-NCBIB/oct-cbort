```python
#Import required system libraries for file management
import sys,importlib,os

# Provide path to oct-cbort library
module_path=os.path.abspath('/Users/damondepaoli/Documents/GitHub/oct-cbort')
if module_path not in sys.path:
    sys.path.append(module_path)

# Import oct-cbort library
from oct import *

# Choose a directory with all meta/ofd/ofb data within it
d = os.path.join(module_path,'examples/data/4_BL_Benchtop_Phantom_struct_angio_ps')
data = Load(directory = d)
```

## All the processing steps outlined in the previous tutorials are handled here within the `Post` processer:

#### There are some options we can select from, but most of the settings are loaded from the reconstructsettings.ini/editsettings.ini files

```python
processer = Post()
processer.processFrameRange(data, startFrame=1, endFrame=1, procState='tomo+angio+struct+ps+stokes', writeState=True,
                          procAll=False, hold=False, holdState='struct', rotCartesian=False)
```

### We can view the real processed images by using the `showProcessedData(data,processer)` function from the utils 

```python
showProcessedData(data,processer)
```
