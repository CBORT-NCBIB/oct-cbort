# Tutorial 9: Editting processing settings using editsetting.ini

```python
#Import required system libraries for file management
import sys,importlib,os

# Provide path to oct-cbort library
module_path=os.path.abspath('/Users/damondepaoli/Documents/GitHub/oct-cbort')
if module_path not in sys.path:
    sys.path.append(module_path)

# Import oct-cbort library
from oct import *
```

First let's look at the results of processing a PS frame without setting anything!

What happens in this case is that the settings are loaded for the .xml file, which have no ps variables/settings. When that occurs, the data/metadata module uses the defaults located within the Metadata class.

```python
d = os.path.join(module_path,'examples/data/4_BL_Benchtop_Phantom_struct_angio_ps')
data = Load(directory = d)
```

Now upon loading that directory, what we don't see is that there is a new file generated labelled "editsettings.ini"

If we open that file and look through, we'll see it mirrors well what we find in the Metadata defaults. 

If we change the values within editsettings.ini, and then **place that file into the /settings sub-directory**, these settings will override the defaults!

The safety-feature of having to move the files into the settings sub-directory is to assure that we only use this settings file when desired. For instance, when using the automated post-processing architecture, the <u>"reconstructsettings.ini"</u> file sent from c++ should be used, and this avoids confusion. 

### Lets prove it

### Case 1: Processing using defaults:

See /oct/load/metadata for defaults

```python
# Process
processer = Post()
processer.processFrameRange(data, startFrame=1, endFrame=1, procState='ps')
```

```python
# Show
showProcessedData(data,processer)
```

### Case 2: Processing using editsettings by dropping it into the /settings directory:

Let's change the following settings from the defaults.

[PsTomSetting]

nZOffset = 5 -> 10

[ProcOptions]

bSpectralBinning = False ->True

```python
# We need to re-load the data 
d = os.path.join(module_path,'examples/data/4_BL_Benchtop_Phantom_struct_angio_ps')
data = Load(directory = d)
```

```python
# Process
processer = Post()
processer.processFrameRange(data, startFrame=1, endFrame=1, procState='ps')
```

```python
# Show
showProcessedData(data,processer)
```
